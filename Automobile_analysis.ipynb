## Predictive Analysis Using Machine Learning Models

### Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
features = ["engine-size", "curb-weight", "width", "length", "horsepower", "highway-mpg", "city-mpg"]
x = df[features]
y = df.price
```

```python
x1_train, x1_test, y1_train, y1_test = train_test_split(x,y,random_state = 1, test_size = 0.3)
reg = LinearRegression()
reg.fit(x1_train, y1_train)
linear_pred = reg.predict(x1_test)
linear_mae = mean_absolute_error(y1_test, linear_pred)
print("Mean Absolute Error for Linear Regression Model is :", linear_mae)
```

```python
print("Shape of the x_train set is",x1_train.shape)
print("Shape of the x_test set is",x1_test.shape)
print("Shape of the y_train set is",y1_train.shape)
print("Shape of the y_test set is",y1_test.shape)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
features = ["engine-size", "curb-weight", "width", "length", "horsepower", "highway-mpg", "city-mpg"]
x = df[features]
y = df.price

train_x2, val_x2, train_y2, val_y2 = train_test_split(x,y, random_state = 0)
ds_model = DecisionTreeRegressor(random_state=0)
ds_model.fit(train_x2, train_y2)
tree_pred = ds_model.predict(val_x2)
tree_mae = mean_absolute_error(val_y2, tree_pred)
print("Mean Absolute Error for Decision Tree Model is :", tree_mae)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
features = ["engine-size", "curb-weight", "width", "length", "horsepower", "highway-mpg", "city-mpg"]
x = df[features]
y = df.price

train_x, val_x, train_y, val_y = train_test_split(x,y, random_state = 1)
ds_model2 = RandomForestRegressor(random_state = 1)
ds_model2.fit(train_x, train_y)
forest_pred = ds_model2.predict(val_x)
mae_forest = mean_absolute_error(val_y, forest_pred)
print("Mean Absolute Error for Random Forest is :",mae_forest)
```

```python
#Including 2 more features to check their significance in predicting the price.
feats = ["horsepower", "engine-size","curb-weight", "width", "length", "highway-mpg", "city-mpg", "wheel-base", "bore"] 
all_feats = df[feats]
all_price = df.price
train_x, val_x, train_y, val_y = train_test_split(all_feats ,all_price)
ds_model2 = RandomForestRegressor()
ds_model2.fit(train_x, train_y)
forest_pred = ds_model2.predict(val_x)
mae_forest = mean_absolute_error(val_y, forest_pred)
mae_forest
```

```python
ds_model2.score(train_x, train_y)
```

train x - all training data
train y - all pred data
val x - testing data
val y - target data
forest_pred - prediction

```python
len(forest_pred)
len(val_y)
```

```python
total_mae = 0
iters = 10
for x in range(iters):
    train_x, val_x, train_y, val_y = train_test_split(all_feats ,all_price)
    ds_model = RandomForestRegressor()
    ds_model.fit(train_x, train_y)
    pred = ds_model.predict(val_x)
    mae_forest = mean_absolute_error(val_y, pred)
    total_mae +=mae_forest
avg_mae = total_mae/iters
print("Avg of",iters,"mae using Random Forest:",avg_mae)
```

Based on the predicted prices using the two models, Random Forest performs better than the Decision Tree and Linear Regression models.
